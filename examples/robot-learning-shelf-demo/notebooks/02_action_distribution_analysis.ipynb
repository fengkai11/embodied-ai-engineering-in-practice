# Action Distribution Analysis
本 notebook 用于读取 `dataset_v0_sample` 中的动作数据，并绘制 dx/dy/dz/gripper_delta 的分布。

```python
from pathlib import Path
import json
import matplotlib.pyplot as plt

dataset_dir = Path('../datasets/dataset_v0_sample')
values = {'dx': [], 'dy': [], 'dz': [], 'gripper_delta': []}
for ep in dataset_dir.iterdir():
    ap = ep / 'actions.jsonl'
    if not ap.exists():
        continue
    for line in ap.read_text(encoding='utf-8').splitlines():
        row = json.loads(line)
        vec = row['delta_ee_xyzrpy']
        values['dx'].append(vec[0])
        values['dy'].append(vec[1])
        values['dz'].append(vec[2])
        values['gripper_delta'].append(row['gripper_delta'])

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
for ax, key in zip(axes.ravel(), ['dx', 'dy', 'dz', 'gripper_delta']):
    ax.hist(values[key], bins=20)
    ax.set_title(f'{key} distribution')
    ax.set_xlabel(key)
    ax.set_ylabel('count')
plt.tight_layout()
plt.show()
```
